```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Raw_customer_orders.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

drop notes as it not needed and adds no value

```python
df.drop('notes',axis=1,inplace=True)
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
df.info()
```

```python
df['total'] = df['quantity'] * df['price']
```

```python
df.info()
```

the 3 customer names wihtout entry is filled with customer_1,customer_2,customer_3 ;

```python
null_customer_index = df[df['customer_name'].isnull()].index
```

```python
df.loc[null_customer_index[0], 'customer_name'] = 'customer_1'
df.loc[null_customer_index[1], 'customer_name'] = 'customer_2'
df.loc[null_customer_index[2], 'customer_name'] = 'customer_3'
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df['email'].unique()
```

```python
df['country'].unique()
```

```python
df['country'].replace('IN','India',inplace=True)
df['country'].replace('india','India',inplace=True)
```

```python
df['payment_status'] = df['payment_status'].str.lower()
```

```python
df['product'] = df['product'].str.lower()
```
